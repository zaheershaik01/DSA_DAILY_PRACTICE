```python
# 1st Answer

def fibonacci(n, memo={}):
    if n in memo:
        return memo[n]
    if n <= 1:
        return n
    memo[n] = fibonacci(n - 1, memo) + fibonacci(n-2, memo)
    return memo[n]

print(fibonacci(10))
print(fibonacci(6))
```

```python
# 2nd Answer

def add_to_dict(a, b=None):
    if b is None:
        b = {}
    b[a] = a**2
    return b

print(add_to_dict(2))
print(add_to_dict(2, {1:1}))
```

```python
# 3rd Answer

def filter_integers(**kwargs):
    return {k: v for k, v in kwargs.items() if isinstance (v, int)}

print(filter_integers(a=1, b='Zaheer', c=25, d=99))
print(filter_integers(x=22, y='shaik', z='Zahy'))
```

```python
# 4th Answer

def apply_callback(callback, lst):
    return [callback(x) for x in lst]

print(apply_callback(lambda x: x**2, [1, 2, 3, 4]))
print(apply_callback(lambda x: x+1, [1, 2, 3, 4]))
```

```python
# 5th Answer

def outer_function():
    def inner_function(x):
        return x ** 2
    return inner_function

square = outer_function()
print(square(4))
print(square(6))
```

```python
# 6th Answer

import time

def timer_decorator(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        print(f"Function {func.__name__} took {end_time - start_time} seconds to execute.")
        return result
    return wrapper

@timer_decorator
def complex_calculation(n):
    return sum(x**2 for x in range(n))

# Test
print(complex_calculation(10000))
```

```python
# 7th Answer

def filter_and_map(filter_func, map_func, lst):
    return [map_func(x) for x in lst if filter_func(x)]

print(filter_and_map(lambda x: x%2==0, lambda x:x**2, [1, 2, 3, 4, 5]))
print(filter_and_map(lambda x: x > 2, lambda x:x+1, [1, 2, 3, 4, 5]))
```

```python
# 8th Answer

def compose(f, g):
    return lambda x: f(g(x))

f = lambda x: x+1
g = lambda x: x*2
h = compose(f, g)
print(h(4))
print(h(5))
```

```python
# 9th Answer

from functools import partial

multiply_by_2 = partial(lambda x, y: x * y, 2)

print(multiply_by_2(5))
print(multiply_by_2(10))
```

```python
# 10th Answer

def average(lst):
    try:
        return sum(lst) / len(lst)
    except ZeroDivisionError:
        return None

print(average([1, 2, 3, 4, 5, 6, 7]))
print(average([]))
```

```python
# 11th Answer

def fibonacci_generator():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a+b

fib_gen = fibonacci_generator()
for _ in range(10):
    print(next(fib_gen))
```

```python
# 12th Answer

def curry_product(x):
    def inner1(y):
        def inner2(z):
            return x * y * z
        return inner2
    return inner1

print(curry_product(1) (2) (3))
print(curry_product(4) (5) (6))
```

```python
# 13th Answer
def write_to_file(lst, filename):
    try:
        with open(filename, 'w') as f:
            for num in lst:
                f.write(f"{num}\n")
    except IOError as e:
        print(f"An error occurred: {e}")

# Test
write_to_file([1, 2, 3, 4, 5], 'output.txt')
```

```python
# 14th Answer

def separate_types(lst):
    ints, strs, floats = [], [], []
    for item in lst:
        if isinstance(item, int):
            ints.append(item)
        elif isinstance(item, str):
            strs.append(item)
        elif isinstance(item, float):
            floats.append(item)
    return ints, strs, floats

print(separate_types([1, 'a', 2.0, 'b', 3, 'c', 5.0, 11, 10.0]))   
```
