```python
# 1: Sum of List Elements

def sum_list(numbers):
    # Your code goes here
    pass
    count = 0
    for num in numbers:
        count += num
        
    return count
```

```python
# 2: Largest Element in a List

def find_largest(numbers):
    # Your code goes here
    pass  
    if not numbers:
        return None
    
    largest = numbers[0]
    for i in numbers[1:]:
        if i > largest:
            largest = i
    
    return largest
```

```python
# 3: Remove Duplicate in a List

def remove_duplicates(lst):
    # Your code goes here
    pass

    unique_list = []
    
    for item in lst:
        if item not in unique_list:
            unique_list.append(item)
            
    return unique_list
```

```python
# 4: Check if all elements in a list are Unique

def check_unique(lst):
    # Your code goes here
    pass

    seen = set()
    for num in lst:
        if num in seen:
            return False
            
        seen.add(num)
    return True
```

```python
# 5: Program to Reverse a List

def reverse_list(lst):
    # Your code goes here
    pass
    
    start = 0
    end = len(lst) -  1 
    
    while start < end:
        lst[start], lst[end] = lst[end], lst[start]
        start += 1 
        end -= 1 
    return lst
```

```python
# 6: Count Number of Odd and Even Elements in a List

def count_even_odd(lst):
    # Your code goes here
    pass

    even_count = 0
    odd_count = 0
    
    for num in lst:
        if num%2==0:
            even_count += 1
        else:
            odd_count += 1 
        
    return even_count, odd_count
```

```python
# 7: Maximum difference between two consecutive elements in a list.

def max_consecutive_difference(lst):
    # Your code goes here
    pass

    if len(lst) < 2:
        return 0
        
    max_diff = 0
    
    for i in range(1, len(lst)):
        curr_diff = abs(lst[i] - lst[i-1])
        
        if curr_diff > max_diff:
            max_diff = curr_diff
            
    return max_diff
```

```python
# 8: Merge two Sorted List

def merge_two_sorted_lists(list1, list2):
    # Your code goes here
    pass

    i, j = 0, 0
    result = []
    
    while i < len(list1) and j < len(list2):
        if list1[i] < list2[j]:
            result.append(list1[i])
            i +=  1 
        else:
            result.append(list2[j])
            j += 1 
            
    while i < len(list1):
        result.append(list1[i])
        i += 1 
        
    while j < len(list2):
        result.append(list2[j])
        j += 1
        
    return result
```

```python
# 9: Rotate a List

def rotate_list(lst, k):
    # Your code goes here
    pass

    if not lst:
        return []
        
    k = k % len(lst)
    
    for _ in range(k):
        last_element = lst.pop()
        lst.insert(0, last_element)
        
    return lst
```

```python
# 10: Merge 2 List into Dictionary

def merge_lists_to_dictionary(keys, values):
    # Your code goes here
    pass
     
    if(len(keys) != len(values)):
        return False
        
    result = {}
    
    for i in range(len(keys)):
        result[keys[i]] = values[i]
        
    return result
```

```python
# 11: Merge Multiple Dictionaries

def merge_three_dictionaries(dict1, dict2, dict3):
    # Your code goes here
    pass
    
    merged_dict = dict1.copy()
    
    for key, value in dict2.items():
        merged_dict[key] = value
        
    for key, value in dict3.items():
        merged_dict[key] = value
        
    return merged_dict
```

```python
# 12: Words Frequency in a Sentence

def count_word_frequency(sentence):
    # Your code goes here
    pass

    word_count = {}
    
    words = sentence.split()
    
    for word in words:
        if word in word_count:
            word_count[word] +=  1 
        else:
            word_count[word] = 1 
    return word_count
```

```python
# 13: Palindromic Tuple

def is_palindromic_tuple(tup):
    # Your code goes here
    pass

    start = 0
    end = len(tup) -  1 
    
    while start <= end:
        if tup[start] != tup[end]:
            return False
            
        start +=  1 
        end -= 1 
        
    return True
```

```python
# 14: Merge Dictionaries with Common Keys

def merge_dicts_with_overlapping_keys(dicts):
    # Your code goes here
    pass

    result = {}
    
    for d in dicts:
        for key, value in d.items():
            if key in result:
                result[key] += value
            else:
                result[key] = value
                
    return result
```

```python
# 15: Check if List is Subset of another List

def is_subset(lst1, lst2):
    # Your code goes here
    pass

    for element in lst1:
        found = False
        
        for item in lst2:
            if item == element:
                found = True
                break
        if not found:
            return False
    return True
```
