```python
# 15th answer

def min_of_tuples(tpl):
    return min(tpl)
def max_of_tuples(tpl):
    return max(tpl)
def sum_of_tuples(tpl):
    return sum(tpl)

sample_tuple = {1, 2, 4, 6, 8, 11, 15}
print(f"Minimum: {min_of_tuples(sample_tuple)}")
print(f"Maximum: {max_of_tuples(sample_tuple)}")
print(f"Sum: {sum_of_tuples(sample_tuple)}")
```
