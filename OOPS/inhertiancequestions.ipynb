```python
## 1st answer

class Animal:
    def __init__(self, name, species):
        self.name = name 
        self.species = species 

class Dog(Animal):
    def __init__(self, name, species, breed):
        super().__init__(name, species)
        self.breed = breed

dog = Dog("Tommy", "Canine", "Golden Retriever")
print(dog.name, dog.species, dog.breed)
```

```python
## 2nd answer

class Dog(Animal):
    def __init__(self, name, species, breed):
        super().__init__(name, species)
        self.breed = breed

    def __str__(self):
        return f"Dog(Name: {dog.name}, Species: {dog.species}, Breed: {dog.species})"

dog = Dog("Tommy", "Canine", "Golden Retriever")
print(dog)
```

```python
## 3rd answer

class Dog(Animal):
    def __init__(self, name, species, breed):
        super().__init__(name, species)
        self.breed = breed

    def bark(self):
        print("Woof! Woof!")

dog = Dog("Tommy", "Canine", "Golden Retriever")
dog.bark()
```

```python
## 4th answer

class Walker:
    def walk(self):
        print("Walking...")
    
class Runner:
    def run(self):
        print("Running...")

class Athlete(Walker, Runner):
        pass 

athlete = Athlete()
athlete.walk()
athlete.run()
```

```python
## 5th answer

class Athlete(Walker, Runner):
    def walk(self):
        print("Athlete Walking...")
        super().walk()

athlete = Athlete()
athlete.walk()
```

```python
## 6th answer

class Athlete(Walker, Runner):
    def __init__(self, training_hours):
        self.training_hours = training_hours

    def train(self):
        print(f"Training for {self.training_hours} hours")

athlete = Athlete(4)
athlete.train()
```

```python
## 7th answer

class A:
    def show(self):
        print("This is the message of A.")

class B(A):
    def show(self):
        print("This is the message of B.")

class C(A):
    def show(self):
        print("This is the message of C.")

class D(B, C):
    pass

d = D()
d.show()
```

```python
## 8th answer

class Shape:
    def __init__(self, color):
        self.color = color

class Circle(Shape):
    def __init__(self, color, radius):
        super().__init__(color)
        self.radius = radius

circle = Circle('Red', 5)
print(circle.color, circle.radius)
```

```python
## 9th answer

class Person:
    def __init__(self, name):
        self.name = name 

class Employee:
    def __init__(self, employee_id):
        self.employee_id = employee_id 

class Manager(Person, Employee):
    def __init__(self, name, employee_id):
        super().__init__(name)
        Employee.__init__(self, employee_id) 

manager = Manager('Zaheer', '22RA1A05I7')
print(manager.name, manager.employee_id)
```

```python
## 10th answer

class Vehicle:
    def start(self):
        print("Vehicle Starting...")

class Car(Vehicle):
    def start(self):
        print("Car Starting...")
        super().start()

car = Car()
car.start()
```

```python
## 11th answer

class flyer:
    def fly(self):
        print("Flying...")

class swimmer:
    def swim(self):
        print("Swimming...")

class Superhero(flyer, swimmer):
    pass 

superhero = Superhero()
superhero.fly()
superhero.swim()
```

```python
## 12th answer

class Base1:
    def __init__(self, a):
        self.a = a 
    
class Base2:
    def __init__(self, b):
        self.b = b
        
class Derived(Base1, Base2):
    def __init__(self, a, b, c):
        super().__init__(a)
        Base2.__init__(self, b)
        self.c = c 

derived = Derived(1, 2, 3)
print(derived.a, derived.b, derived.c) 
```

```python
## 13th answer

class Animal:
    pass 

class Cat(Animal):
    pass 

cat = Cat()
animal = Animal()
print(isinstance(animal, Animal))
print(isinstance(cat, Cat))
print(isinstance(cat, Animal))
print(isinstance(animal, Cat))
```

```python
## 14th answer

class Bird:
    def speak(self):
        pass 

class Parrot(Bird):
    def speak(self):
        print("Parrot says: Quawk!")

class Penguin(Bird):
    def speak(self):
        print("Penguin says: Honk!")

birds = [Parrot(), Penguin()]
for bird in birds:
    bird.speak()
```

```python
## 15th answer

class Device:
    def __init__(self, brand):
        self.brand = brand 

class Phone(Device):
    def __init__(self, model, brand):
        super().__init__(brand)
        self.model = model 

class Camera:
    def __init__(self, resolution):
        self.resolution = resolution 

class Smartphone(Phone, Camera):
    def __init__(self, brand, model, resolution):
        Phone.__init__(self, brand, model)
        Camera.__init__(self, resolution)

smartphone = Smartphone('Apple', 'iPhone 12', '12 MP')
print(smartphone.brand, smartphone.model, smartphone.resolution)
```
